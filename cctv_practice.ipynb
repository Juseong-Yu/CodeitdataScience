```python
import os
import requests
import json
import pandas as pd
import seaborn as sns
```

```python
file_type = 'json'
service_name = 'BukChonInOutPeopleInfo'
key = '67434745776161613130336e6e544368'
url = "http://openapi.seoul.go.kr:8088/%s/%s/%s/1/100/" %(key,file_type,service_name)
url
```

```python
url ="http://openapi.seoul.go.kr:8088/%s/%s/%s/1/1000/" %(key,file_type,service_name)
path = os.path.join(url) 
path
```

```python
response = requests.get(path)
texter = response.text
texter
```

```python
dict_data = json.loads(texter)
dict_data
```

```python
type(dict_data)
```

```python
print(dict_data.keys())
```

```python
import pandas as pd
df = pd.DataFrame(dict_data['BukChonInOutPeopleInfo']['row'])
df
```

```python
MONTH = []
for data in df['ENDTIME']:
  MONTH.append(data.split('-')[1])
df['MONTH'] = MONTH
df
```

```python
nov_df = df[df['MONTH'] == '11']
nov_df['INCOUNT'].mean()
```

```python
oct_df = df[df['MONTH']=='10']
print(oct_df['OUTCOUNT'].mean())
print(nov_df['OUTCOUNT'].mean())
```

```python
print(oct_df['OUTCOUNT'].max())
print(oct_df['OUTCOUNT'].min())
print()
print(oct_df['INCOUNT'].max())
print(oct_df['INCOUNT'].min())
print()
print(nov_df['OUTCOUNT'].max())
print(nov_df['OUTCOUNT'].min())
print()
print(nov_df['INCOUNT'].max())
print(nov_df['INCOUNT'].min())
```
